```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train_df=pd.read_excel('flight_train.xlsx')
test_df=pd.read_excel('flight_test.xlsx')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# combine train and test df
df = pd.concat([train_df, test_df])
df.head()
```

```python
df.info()
```

```python
# feature engineering process
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(":").str[0].astype(int)
df['Arrival_Minutes']=df['Arrival_Time'].str.split(":").str[1].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dept_Hour']=df['Dep_Time'].str.split(":").str[0].astype(int)
df['Dept_Minutes']=df['Dep_Time'].str.split(":").str[1].astype(int)
```

```python
df.head()
```

```python
df['Dept_Hour']= df['Dept_Hour'].astype(int)
df['Dept_Minutes']=df['Dept_Minutes'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,"3 stops":3,"4 stops":4,"nan":1})
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
duration_hr=df['Duration'].str.split('h').str[0].astype(int)
duration_hr
```

```python
duration_min=df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0).astype(int)
duration_min
```

```python
df.drop(6474,axis=0,inplace=True)
df.drop(2660,axis=0,inplace=True)
```

```python
df.info()
```

```python
df['Total_Time'] = duration_hr*60 + duration_min
```

```python
df.info()
```

```python
df.drop('Duration',axis=1, inplace=True)
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
```
